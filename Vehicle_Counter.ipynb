```
#Making all Autos as class 80

import glob

dataset_root = "/content/drive/MyDrive/auto_dataset"

label_dirs=[
    f"{dataset_root}/train/labels",
    f"{dataset_root}/valid/labels",
    f"{dataset_root}/test/labels",
]

for d in label_dirs:
    if not os.path.exists(d):
      continue
    print("Fixing labels in", d)
    for path in glob.glob(f"{d}/*.txt"):
      new_lines = []
      with open(path, "r") as f:
        for line in f:
          line=line.strip()
          if not line:
            continue
          parts=line.split()
          parts[0]="80"
          new_lines.append(" ".join(parts))
      with open(path, "w") as f:
        f.write("\n".join(new_lines))
```

```
import cv2
from ultralytics import YOLO
import numpy as np
import argparse
import os
import pandas as pd
from typing import Tuple

# ---------------- CONFIG ----------------
FRAME_SKIP = 10

CUSTOM_MODEL_PATH = "/content/drive/MyDrive/Vehicle/auto_best.pt"
GENERAL_MODEL_PATH = "yolov8n.pt"


LINE_X_PERCENTAGE = 0.5

TARGET_WIDTH = 1280
CONFIDENCE_THRESHOLD = 0.3


VEHICLE_CLASS_IDS = [1, 2, 3, 5, 7, 80]

REPORT_PATH = "/content/drive/MyDrive/Vehicle/traffic_report.xlsx"


def save_to_excel(counts: dict, total_count: int, video_path: str):
    """
    Append one row to an Excel file without duplicating headers.
    If file exists â†’ append row only.
    If not â†’ create file with headers once.
    """
    row = {
        "video": os.path.basename(video_path),
        "total": total_count,
        "car": counts.get("car", 0),
        "motorcycle": counts.get("motorcycle", 0),
        "bus": counts.get("bus", 0),
        "truck": counts.get("truck", 0),
        "autorickshaw": counts.get("autorickshaw", 0),
        "bicycle": counts.get("bicycle", 0)
    }

    df_new = pd.DataFrame([row])

    if not os.path.exists(REPORT_PATH):
        df_new.to_excel(REPORT_PATH, index=False)
        print(f"Created new report file: {REPORT_PATH}")
    else:

        df_existing = pd.read_excel(REPORT_PATH)

        for col in df_existing.columns:
            if col not in df_new.columns:
                df_new[col] = None

        cols = list(df_existing.columns) + [c for c in df_new.columns if c not in df_existing.columns]
        df_all = pd.concat([df_existing, df_new[cols]], ignore_index=True)
        df_all.to_excel(REPORT_PATH, index=False)
        print(f"Appended new row to: {REPORT_PATH}")


def iou_xyxy(boxA, boxB) -> float:
    xA = max(boxA[0], boxB[0])
    yA = max(boxA[1], boxB[1])
    xB = min(boxA[2], boxB[2])
    yB = min(boxA[3], boxB[3])
    interW = max(0, xB - xA)
    interH = max(0, yB - yA)
    interArea = interW * interH
    boxAArea = max(0, boxA[2] - boxA[0]) * max(0, boxA[3] - boxA[1])
    boxBArea = max(0, boxB[2] - boxB[0]) * max(0, boxB[3] - boxB[1])
    denom = (boxAArea + boxBArea - interArea)
    return interArea / denom if denom > 0 else 0.0


def safe_extract_boxes_ids_cls(result):
    if not result or len(result) == 0:
        return np.empty((0, 4), dtype=int), np.array([], dtype=int), np.array([], dtype=int)
    r = result[0]
    boxes_attr = getattr(r, "boxes", None)
    if boxes_attr is None:
        return np.empty((0, 4), dtype=int), np.array([], dtype=int), np.array([], dtype=int)

    try:
        ids_attr = getattr(boxes_attr, "id", None)
        xyxy = boxes_attr.xyxy.cpu().numpy() if getattr(boxes_attr, "xyxy", None) is not None else np.empty((0,4))
        if ids_attr is None:
            return np.empty((0, 4), dtype=int), np.array([], dtype=int), np.array([], dtype=int)
        ids = boxes_attr.id.cpu().numpy().astype(int)
        cls = boxes_attr.cls.cpu().numpy().astype(int)
        return xyxy.astype(int), ids.astype(int), cls.astype(int)
    except Exception:
        return np.empty((0, 4), dtype=int), np.array([], dtype=int), np.array([], dtype=int)


def map_name_to_key(name: str) -> str:

    if not name:
        return None
    name = name.lower()
    if name == "car":
        return "car"
    if name in ["motorcycle", "motorbike", "bike"]:
        return "motorcycle"
    if name == "bus":
        return "bus"
    if name == "truck":
        return "truck"
    if name == "autorickshaw" or name == "auto" or "rickshaw" in name:
        return "autorickshaw"
    if name == "bicycle" or name == "bike":
        return "bicycle"
    return None


def get_label_name(cls_id: int, general_names, custom_names) -> str:

    name = None
    try:
        if isinstance(general_names, dict):
            name = general_names.get(cls_id)
        else:
            name = general_names[cls_id] if cls_id < len(general_names) else None
    except Exception:
        name = None

    if name:
        return name

    try:
        if isinstance(custom_names, dict):
            name = custom_names.get(cls_id)
        else:
            name = custom_names[cls_id] if cls_id < len(custom_names) else None
    except Exception:
        name = None

    return name


def main():
    parser = argparse.ArgumentParser()
    parser.add_argument("--video", default="/content/drive/MyDrive/Vehicle/daily.mp4", type=str)
    parser.add_argument("--output", default="/content/drive/MyDrive/Vehicle/output_merged.mp4", type=str)
    parser.add_argument("--custom", default=CUSTOM_MODEL_PATH, type=str, help="path to custom autorickshaw model")
    parser.add_argument("--general", default=GENERAL_MODEL_PATH, type=str, help="path to general yolov8 model (e.g., yolov8n.pt)")
    parser.add_argument("--frame_skip", default=FRAME_SKIP, type=int)
    parser.add_argument("--conf", default=CONFIDENCE_THRESHOLD, type=float)
    args, _ = parser.parse_known_args()

    FRAME_SKIP_LOCAL = args.frame_skip
    CONF = args.conf

    # Load models
    print(f"Loading custom model: {args.custom}")
    try:
        custom_model = YOLO(args.custom)
    except Exception as e:
        print(f"Error loading custom model ({args.custom}): {e}")
        return

    print(f"Loading general model: {args.general}")
    try:
        general_model = YOLO(args.general)
    except Exception as e:
        print(f"Error loading general model ({args.general}): {e}")
        return

    # Retrieve name maps
    general_names = getattr(general_model, "names", None)
    custom_names = getattr(custom_model, "names", None)
    print("General model classes sample:", list(general_names.items())[:10] if isinstance(general_names, dict) else (general_names[:10] if general_names else "unknown"))
    print("Custom model classes sample:", custom_names)

    # Video capture
    cap = cv2.VideoCapture(args.video)
    if not cap.isOpened():
        print(f"Error: Cannot open video {args.video}")
        return

    original_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    original_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = cap.get(cv2.CAP_PROP_FPS)
    fps = int(fps) if fps and fps > 0 else 25

    scale_factor = TARGET_WIDTH / original_width
    target_height = int(original_height * scale_factor)
    roi_width = int(TARGET_WIDTH * LINE_X_PERCENTAGE)

    # Video Writer
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    writer = cv2.VideoWriter(args.output, fourcc, fps, (TARGET_WIDTH, target_height))

    # Counters
    counts = {"car": 0, "motorcycle": 0, "bus": 0, "truck": 0, "autorickshaw": 0, "bicycle": 0}
    total_vehicles = 0
    previous_objects = {}
    counted_ids = set()

    frame_count = 0
    retries = 0
    MAX_RETRIES = 5

    print(f"Processing started. Using {LINE_X_PERCENTAGE*100:.0f}% of video width (ROI width = {roi_width}px).")
    IOU_MATCH_THRESH = 0.4
    next_synthetic_id = 1_000_000

    while cap.isOpened():
        success, frame = cap.read()

        if not success:
            retries += 1
            if retries >= MAX_RETRIES:
                print("\nEnd of video (or too many read errors).")
                break
            continue
        retries = 0

        frame = cv2.resize(frame, (TARGET_WIDTH, target_height))
        frame_count += 1

        roi_frame = frame[0:target_height, 0:roi_width]
        current_objects = {}

        if frame_count % FRAME_SKIP_LOCAL == 0:
            try:
                res_general = general_model.track(roi_frame, persist=True, conf=CONF, verbose=False)
            except Exception as e:
                print(f"Warning: general_model.track() failed on frame {frame_count}: {e}")
                res_general = None
            try:
                res_custom = custom_model.track(roi_frame, persist=True, conf=CONF, verbose=False)
            except Exception as e:
                print(f"Warning: custom_model.track() failed on frame {frame_count}: {e}")
                res_custom = None


            gen_boxes, gen_ids, gen_cls = safe_extract_boxes_ids_cls(res_general)
            cust_boxes, cust_ids, cust_cls = safe_extract_boxes_ids_cls(res_custom)


            merged_boxes = []
            merged_ids = []
            merged_cls = []

            for b, tid, cid in zip(gen_boxes, gen_ids, gen_cls):
                merged_boxes.append(b.astype(int))
                merged_ids.append(int(tid))
                merged_cls.append(int(cid))


            for cb, ctid, ccid in zip(cust_boxes, cust_ids, cust_cls):
                cb = cb.astype(int)
                best_iou = 0.0
                best_idx = -1
                for idx, gb in enumerate(merged_boxes):
                    i = iou_xyxy(cb, gb)
                    if i > best_iou:
                        best_iou = i
                        best_idx = idx
                if best_iou >= IOU_MATCH_THRESH and best_idx >= 0:

                    merged_cls[best_idx] = int(ccid)
                else:
                    merged_boxes.append(cb)
                    merged_ids.append(next_synthetic_id)
                    merged_cls.append(int(ccid))
                    next_synthetic_id += 1

            # Convert to numpy arrays for iteration
            if len(merged_boxes) > 0:
                boxes = np.array(merged_boxes)
                track_ids = np.array(merged_ids).astype(int)
                cls_ids = np.array(merged_cls).astype(int)
            else:
                boxes = np.empty((0,4), dtype=int)
                track_ids = np.array([], dtype=int)
                cls_ids = np.array([], dtype=int)


            for box, track_id, cls_id in zip(boxes, track_ids, cls_ids):
                x1, y1, x2, y2 = box.astype(int)
                center_x = (x1 + x2) // 2


                name = get_label_name(int(cls_id), general_names, custom_names)
                if name is None:
                    name = str(cls_id)

                keyname = map_name_to_key(name)
                if keyname is None:
                    continue


                current_objects[int(track_id)] = {
                    'class_id': int(cls_id),
                    'last_x': int(center_x),
                    'name': name
                }


                x1c, y1c = max(0, x1), max(0, y1)
                x2c, y2c = min(roi_width-1, x2), min(target_height-1, y2)
                cv2.rectangle(frame, (x1c, y1c), (x2c, y2c), (0, 255, 255), 2)
                cv2.putText(frame, f"{name}", (x1c, max(10, y1c-6)),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255,255,255), 1)


            lost_ids = set(previous_objects.keys()) - set(current_objects.keys())
            for track_id in lost_ids:
                if track_id in counted_ids:
                    continue

                obj_data = previous_objects[track_id]
                name = obj_data.get('name')
                key = map_name_to_key(name)

                if key is not None:
                    counts[key] = counts.get(key, 0) + 1
                    total_vehicles += 1
                    counted_ids.add(track_id)
                    print(f"â• Counted {name} (ID {track_id})")


            previous_objects = current_objects.copy()


        stats = (f"Total:{total_vehicles} | Car:{counts['car']} "
                 f"Bike:{counts['motorcycle']} Bus:{counts['bus']} "
                 f"Truck:{counts['truck']} Auto:{counts['autorickshaw']} "
                 f"Bicycle:{counts['bicycle']}")
        cv2.putText(frame, stats, (20, 40),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 2)

        writer.write(frame)

        if frame_count % 30 == 0:
            print(f"Processed Frame: {frame_count}", end="\r")


    print("\nVideo ended. Counting vehicles still on screen (remaining tracks)...")
    for track_id, obj_data in previous_objects.items():
        if track_id not in counted_ids:
            name = obj_data.get('name')
            key = map_name_to_key(name)
            if key is not None:
                counts[key] = counts.get(key, 0) + 1
                total_vehicles += 1
                counted_ids.add(track_id)
                print(f"â• Counted remaining {name} (ID {track_id})")

    cap.release()
    writer.release()
    print("\nProcessing Complete.")
    print("Final counts:", counts, "Total:", total_vehicles)

    save_to_excel(counts, total_vehicles, args.video)


if __name__ == "__main__":
    main()
```
